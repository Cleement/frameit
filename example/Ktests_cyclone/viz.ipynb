```python
from pathlib import Path
from frameit.core.settings_class import SimulationConfig
from frameit.core.runner import FrameitRunner

import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import xarray as xr
import math

conf = SimulationConfig.from_yaml_with_model_preset(Path("configs/arome_oper_batsirai.yml"))
# conf = SimulationConfig.from_yaml_with_model_preset(Path("configs/arome_oper_belna.yml"))

# conf = SimulationConfig.from_yaml_with_model_preset(Path("configs_ktests/mnh_chido.yml"))
# conf = SimulationConfig.from_yaml_with_model_preset(Path("configs/mnh_ianos.yml"))

conf.printID()

runner = FrameitRunner(conf)
runner.run()
```

## MesoNH

```python
runner.dict_polar_user['level']
```

```python
from frameit.viz.polar_axes import setup_geodesic_polar_ax
tt=1 ; lev=10
fig2, ax2 = plt.subplots(2,2, figsize=(10,10),subplot_kw={'projection': 'polar'})
setup_geodesic_polar_ax(ax2)
runner.dict_polar_user['level'].vrad.isel(time=tt,level=lev).plot.contourf(ax=ax2[0,0])
runner.dict_polar_user['level'].vtan.isel(time=tt,level=lev).plot.contourf(ax=ax2[0,1])
lev=5
runner.dict_polar_user['level'].vrad.isel(time=tt,level=lev).plot.contourf(ax=ax2[1,0])
runner.dict_polar_user['level'].vtan.isel(time=tt,level=lev).plot.contourf(ax=ax2[1,1])
                                                                                                                                
plt.tight_layout() 
```

```python
fig2, ax2 = plt.subplots(2,2, figsize=(10,10),)
tt=1 
runner.dict_polar_user['level'].vrad.isel(time=tt).mean(['theta']).plot.contourf(ax=ax2[0,0])
runner.dict_polar_user['level'].vtan.isel(time=tt).mean(['theta']).plot.contourf(ax=ax2[0,1])
tt=3
runner.dict_polar_user['level'].vrad.isel(time=tt).mean(['theta']).plot.contourf(ax=ax2[1,0])
runner.dict_polar_user['level'].vtan.isel(time=tt).mean(['theta']).plot.contourf(ax=ax2[1,1])
                                                                                                                                
plt.tight_layout() 
```
